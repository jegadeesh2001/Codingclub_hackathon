```python
dist_group=dist_group.drop('YEAR',axis=1)
columns=list(dist_group.columns)
dist_group['Crime Index']=0
```

```python
for i in range(len(dist_group)):
  sum=0
  for j in range(20):
    # print(df.iloc[i][columns[j]],",",importance_scores[j])
    sum+=dist_group.iloc[i][columns[j]]*importance_scores[j]
  print(sum)
  dist_group.iloc[i,dist_group.columns.get_loc('Crime Index')]=sum
```
